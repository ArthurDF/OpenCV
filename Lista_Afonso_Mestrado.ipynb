```python
import numpy as np
import scipy as sp
from scipy import stats
from scipy.stats import uniform
import matplotlib.pyplot as plt


n1=10000

x = []
x.append(uniform.rvs(size=n1, loc=0, scale=1))

for n in range(1,12):
  x.append(x[n-1]+uniform.rvs(size=n1, loc=0, scale=1))

mean_x = []
for n in range(0,12):
  mean_x.append(x[n].mean())

var_x = []
for n in range(12):
  var_x.append(np.var(x[n]))

sec_mom_x = []
for n in range(12):
  sec_mom_x.append(np.mean(np.power(x[n],2)))

cdf = []
for n in range(12):
  cdf.append(stats.norm.cdf(x[n]))

for n in range(12):
  print("n:",n+1)
  print("média:",mean_x[n])
  print("variância",var_x[n])
  print("segundo momento:",sec_mom_x[n])
  print("f(x)")
  plt.hist(x[n], density=True, bins=30)  
  plt.ylabel('Probability')
  plt.xlabel('Data');
  plt.show()
  print("F(x)")
  plt.hist(x[n], cumulative=True, label='CDF',bins=1000)
  plt.show()

  
  print("_____________________________")
```

É possível se observar que quanto a soma das distribuições uniformes tenda a distribuição normal a medida que o somatório aumenta.

Isso ocorre devido a aproximação da expressão do somatório das distribuições uniformes com a expressão da distribuição normal.

> E(n)=n/2

A variância em função de n pode ser dada como com b = 1 e a=0

```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np

X= stats.uniform.rvs(loc=0,scale=1,size=1000)
a = [0, 0.5, 1, 2]
b= [-1, 0, 1, 2, 4]
for i in a:
  for j in b:
    Y=i*(X**2)+j
    plt.hist(Y,density=True, bins = int(np.sqrt(len(Y))))
    plt.show()
    print("Resultados para a =",i," e b =",j)
    print("E[X] =",X.mean())
    print("E[Y] =",Y.mean())
    print("E[X²] =",(X**2).mean())
    print("E[Y²] =",(Y**2).mean())
    print("Var(X) =",X.var())
    print("Var(Y) =",Y.var())
    print("s(X) =",X.std())
    print("s(Y) =",Y.std())
    print("Corr(X,Y) =",np.corrcoef(X,Y))
    print("r(X,Y) =",stats.pearsonr(X,Y))
```

Nessa questão, para o valor de a=0, temos que o coeficiente de correlação apresenta saídas nan como resultado. Isso se dá porquê nesse cenário não há variação em um dos vetores, que se mostra como uma lista de constantes (Y), dessa forma, como o desvio padrão é zero, o coeficiente tem como resultado uma indeterminação . Ademais, percebe-se que o coeficiente de correlação apresenta valores praticamente iguais para todos os valores de a e b, se a é diferente de zero.

```python
import numpy as np
import scipy as sp
from scipy import stats
from scipy.stats import uniform, expon, norm
import matplotlib.pyplot as plt

#definida a variância e média de X e Y teóricos
mean_x = 1
mean_y=3
var_x=1
var_y=4
x_range = np.linspace(-3, 4, 500)
y_range = np.linspace(-4, 10, 500)
def normpdf(x, mean, sd):
    var = float(sd)**2
    denom = (2*np.pi*var)**.5
    num = np.exp(-(x-mean)**2/(2*var))
    return num/denom

x_teorico = normpdf(x_range, mean_x, var_x)

y_teorico = normpdf(y_range, mean_y, np.sqrt(var_y))
z_teorico = normpdf(y_range, mean_y+mean_x, np.sqrt(var_y+var_x))

x = norm.rvs(size=5000,loc=1)
plt.plot(x_range,x_teorico)
plt.hist(x, density=True, bins=30) 
plt.show()
y = norm.rvs(size=5000, loc =3 , scale=2)

z = x+y
print("x teórico e x amostrado")
plt.plot(x_range,x_teorico)
plt.hist(x, density=True, bins=30) 
plt.show()
print("y teórico e y amostrado")
plt.plot(y_range,y_teorico)
plt.hist(y, density=True, bins=30) 
plt.show()
print("z teórico e z amostrado")
plt.plot(y_range,z_teorico)
plt.hist(z, density=True, bins=30) 
plt.show()

a = np.stack((x, y), axis=0)

print("E(x):",x.mean())
print("E_Teorica(x):",mean_x)
print("E(y):",y.mean())
print("E_Teorica(y):",mean_y)
print("E(z):",z.mean())
print("E_Teorica(z):",mean_x+mean_y)
print("var(x):",x.var())
print("var_teorica(x):",var_x)
print("var(y):",y.var())
print("var_teorica(y):",var_y)
print("var(z):",z.var())
print("var_teorica(z):",var_x+var_y)
print("Cov(X,Y):",np.cov(a))
```

```python
#função teórico para distribuição univariada
def univariate_normal(x, mean, variance):
    """pdf of the univariate normal distribution."""
    return ((1. / np.sqrt(2 * np.pi * variance)) * 
            np.exp(-(x - mean)**2 / (2 * variance)))
#separar valores da matriz de covariância
C = Sigma[0, 1]  # = C transpose in this case
A = Sigma[0, 0]
B = Sigma[1, 1]
# Calculate y|x

#calculo da média e da covariância de Y/X=2
x_condition = 2.  # To condition on x
mean_ygivenx = mean_y + (C * (1/A) * (x_condition - 5))
cov_ygivenx = B - (C * (1/A) * C)

#gerar resultado teórico
yx = np.linspace(-5, 5, num=50)
pyx = univariate_normal(yx, mean_ygivenx, cov_ygivenx)

#gerar distribuição prática
y = norm.rvs(size=5000, loc =mean_ygivenx , scale=cov_ygivenx)

# Plot teórico e prático
plt.plot(yx, pyx, 'r--', 
         label=f'$p(y|x={x_condition:.1f})$')
plt.legend(loc=0)
plt.hist(y, density=True, bins=30) 
plt.xlabel('density', fontsize=13)
plt.xlim(-1.5, 7)
plt.show()
```
